# Listas

```python
# Colocando mis elementos en []

my_list_1 = [3, 5, 2]
print(my_list_1)

my_list_2 = []
print(my_list_2)

my_list_3 = [3, 'Hola', 44.4]
print(my_list_3)

# lista anidada
my_list_4 = ["apple", [3, 'Hola', 44.4], 38]
print(my_list_4)
```

```python
my_list_4.insert(1, 30)
```

```python
print(my_list_4)
```

```python
my_list = ['x', 'l', 'm', 'n']

# Primer elemento
print(my_list[0])

# Tercer elemento
print(my_list[2])

# Ultimo elemento
print(my_list[3])
```

```python
# lista Anidad
n_list = ['Happy', [6, 2, 1], 4]
print(n_list)

print(n_list[0][2])

print(n_list[1][2])

n_list[0] = 'r'

print(n_list)
```
